```python
rests = pd.read_csv("locals.csv", index_col=0)
rests.index += 1
rests = rests.reset_index()
rests.rename(columns={'index': 'restaurant_id'}, inplace=True)
rests.to_excel("restaurants.xlsx")
rests
```

```python
companies = rests
companies = companies[["company"]]
companies = companies.drop_duplicates()
companies = companies.reset_index()
companies.drop('index', axis=1, inplace=True)
companies.index += 1
companies = companies.reset_index()
companies.rename(columns={'index': 'company_id'}, inplace=True)
companies.to_excel("companies.xlsx")
companies
```

```python
dbName = "reviews"
```

```python
connectionData = f"mysql+pymysql://root:{password}@localhost/{dbName}"
engine = create_engine(connectionData)

# Add the DataFrame to the database
all.to_sql("main", con=engine, if_exists='replace', index=False)

rests.to_sql("restaurants", con=engine, if_exists='replace', index=False)

companies.to_sql("companies", con=engine, if_exists='replace', index=False)
```

```python
query = """

SELECT restaurant, AVG(rating) AS rating
FROM main
WHERE chronological < 100
GROUP BY restaurant
ORDER BY rating DESC

"""

pd.read_sql_query(query, engine)
```

```python
query = """

SELECT companies.company, AVG(main.rating) AS rating
FROM companies
JOIN main ON companies.company_id = main.company_id
GROUP BY companies.company
ORDER BY rating DESC
"""

pd.read_sql_query(query, engine)
```

```python
query = """

SELECT companies.company, main.rating, COUNT(main.rating) AS count
FROM companies
JOIN main ON companies.company_id = main.company_id
GROUP BY companies.company, main.rating
ORDER BY rating
"""

pd.read_sql_query(query, engine)
```
